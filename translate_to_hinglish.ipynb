```python
from transformers import MarianTokenizer, MarianMTModel

# Load the pre-trained model and tokenizer
model_name = "Helsinki-NLP/opus-mt-en-ROMANCE"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
from transformers import MarianTokenizer, MarianMTModel

# Load the pre-trained model and tokenizer
model_name = "Helsinki-NLP/opus-mt-en-ROMANCE"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
def translate_to_hinglish(text, model, tokenizer):
    inputs = tokenizer.encode(">>en<<" + text, return_tensors="pt")
    outputs = model.generate(inputs, max_length=50, num_beams=5, early_stopping=True)
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return translated_text

# Example usage
sentences = [
    "Definitely share your feedback in the comment section.",
    "So even if it's a big video, I will clearly mention all the products.",
    "I was waiting for my bag."
]

for sentence in sentences:
    translated_sentence = translate_to_hinglish(sentence, model, tokenizer)
    print("Input English:", sentence)
    print("Output Hinglish:", translated_sentence)
```
